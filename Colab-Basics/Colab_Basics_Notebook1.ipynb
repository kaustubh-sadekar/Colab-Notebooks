## Reading an image form google drive using OpenCV and displaying using matplotlib

You can create a different folder in google drive where you store your images and videos for experiments.

```
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import cv2

#Absolute path of the image saved in drive.
path = "/content/drive/My Drive/colab_data/image1.jpg"
img = cv2.imread(path)
plt.imshow(img)
plt.show()
```

#### Something is wrong with the above image!!!! 
This is because `opencv uses BGR` format while storing the image after reading it where as other than opencv the format used is RGB thus we need to convert the color channesl of the image from BGR to RGB.

```
img = cv2.imread("/content/drive/My Drive/colab_data/image1.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)   # That's the magical line !!!!!
plt.imshow(img)
plt.show()
```

## Dynamic plotting using matplotlib

```
%matplotlib inline
import time
import pylab as pl
from IPython import display
for i in range(10):
    pl.plot(pl.randn(100))
    display.clear_output(wait=True)
    display.display(pl.show())
    time.sleep(1)
```

```
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1) 

for i in range(20):
    x = np.arange(0, i, 0.1);
    y = np.sin(x)
    
    ax.set_xlim(0, i)
    
    ax.cla()
    ax.plot(x, y)
    display(fig)
    
    clear_output(wait = True)
    plt.pause(0.05)
```

```
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1) 

for i in range(21):
    ax.set_xlim(0, 20)
    
    ax.plot(i, 1,marker='x')
    display(fig)
    
    clear_output(wait = True)
    plt.pause(0.5)
```
