# **Basic global thresholding methods using OpenCV and Numpy**
There are several types of thresholding. We will try to impliment some of the thresholding methods using OpenCV and Numpy.

```
# Run this block to files needed for this tutorial
!git clone https://github.com/kaustubh-sadekar/DataFolder.git
```

```
%matplotlib inline
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

#### Importing and displaying the input image

```
img_path = '/content/DataFolder/colab-images/image.jpg'
image = cv2.imread(img_path,0) # Directly reading image as a gray scale image

plt.figure(num=None,figsize=(16,12),dpi=80)
plt.imshow(image,cmap="gray")
plt.show()
```

## **Basic threshold method (cv2.THRESH_BINARY)**

Here we compare each pixel of the gray image with a threshold and if it is greater than the threshold it is set to 255 else it is set to 0, giving us a binary image. This is for cv2.THRESH_BINARY method. In cv2.THRESH_BINARY_INV we perform same operation if the pixel value is less than the threshold.

```
# First using opencv method
threshold_value = 127
ret,thresh1 = cv2.threshold(image,threshold_value,255,cv2.THRESH_BINARY)
ret,thresh1_inv = cv2.threshold(image,threshold_value,255,cv2.THRESH_BINARY_INV)

# Using numpy method
np_thresh1 = (image > threshold_value)
np_thresh1_inv = image < threshold_value

f = plt.figure(num=None, figsize=(16,12), dpi=80)
ax1 = f.add_subplot(2,2, 1)
plt.imshow(thresh1,cmap="gray")
ax1.title.set_text('OpenCV threshold')
ax2 = f.add_subplot(2,2, 2)
plt.imshow(thresh1_inv,cmap="gray")
ax2.title.set_text('OpenCV Inverse threshold')
ax3 = f.add_subplot(2,2, 3)
plt.imshow(np_thresh1,cmap="gray")
ax3.title.set_text('Numpy threshold')
ax4 = f.add_subplot(2,2, 4)
plt.imshow(np_thresh1_inv,cmap="gray")
ax4.title.set_text('Numpy Inverse threshold')
plt.show()
```

## **Truncate thresholding**
In this type of thresholding if the gray value is above threshold then it is clipped to the threshold value else it is kept unchanged.

```
# First using opencv method
threshold_value = 170
ret,trunc = cv2.threshold(image,threshold_value,255,cv2.THRESH_TRUNC)

# Using numpy method
np_trunc = image.copy()
np_trunc[(image > threshold_value)] = threshold_value

f = plt.figure(num=None, figsize=(16,12), dpi=80)
ax1 = f.add_subplot(1,2, 1)
plt.imshow(trunc,cmap="gray")
ax1.title.set_text('OpenCV Truncated threshold')
ax2 = f.add_subplot(1,2, 2)
plt.imshow(np_trunc,cmap="gray")
ax2.title.set_text('Numpy Truncated threshold')
plt.show()
```

## **Threshold to Zero and Threshold to Zero Inverse**

In this method the pixel values less than threshold are set to zero and in threshold to zero inverse the pixels with value greater than threshold are set to zero

```
# First using opencv method
threshold_value = 127
ret,thresh1 = cv2.threshold(image,threshold_value,255,cv2.THRESH_TOZERO)
ret,thresh1_inv = cv2.threshold(image,threshold_value,255,cv2.THRESH_TOZERO_INV)

# Using numpy method
np_thresh1 = image.copy()
np_thresh1[(image < threshold_value)] = 0
np_thresh1_inv = image.copy()
np_thresh1_inv[(image > threshold_value)] = 0

f = plt.figure(num=None, figsize=(16,12), dpi=80)
ax1 = f.add_subplot(2,2, 1)
plt.imshow(thresh1,cmap="gray")
ax1.title.set_text('OpenCV threshold')
ax2 = f.add_subplot(2,2, 2)
plt.imshow(thresh1_inv,cmap="gray")
ax2.title.set_text('OpenCV Inverse threshold')
ax3 = f.add_subplot(2,2, 3)
plt.imshow(np_thresh1,cmap="gray")
ax3.title.set_text('Numpy threshold')
ax4 = f.add_subplot(2,2, 4)
plt.imshow(np_thresh1_inv,cmap="gray")
ax4.title.set_text('Numpy Inverse threshold')
plt.show()
```
